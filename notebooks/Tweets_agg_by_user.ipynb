```python
# Import libraries

import pandas as pd
import os
```

```python
# Load raw data from scraping

tweets_clean = pd.read_csv('../data/tweets_clean.csv')
tweets_clean.head()
```

```python
tweets_clean['name'].info()
```

```python
tweets_clean['name'].info()
```

```python
# Agg

tweets_agg = tweets_clean.groupby('name').agg({'id': 'count', 
                                  'replyCount': ['sum', 'mean'],
                                  'retweetCount': ['sum', 'mean'],
                                  'likeCount': ['sum', 'mean'],
                                  'quoteCount': ['sum', 'mean']
                                 })

tweets_agg.columns = tweets_agg.columns.droplevel()

tweets_agg.rename(columns={tweets_agg.columns[0]: "Total Tweets"})
```
