```python
# Import packages

import os
import pandas as pd
from tqdm import tqdm 
```

```python
# Initialize variables
# Selected users
techCEOs = ['elonmusk', 'sundarpichai', 'tim_cook', 'finkd', 'JeffBezos', 'satyanadella', 'BillGates']
politicians = ['ZelenskyyUa', 'POTUS', 'AOC', 'GovRonDeSantis', 'BarackObama', 'narendramodi'] #'MedvedevRussiaE', 'MedvedevRussia']
celebs = ['KimKardashian', 'rihanna', 'justinbieber', 'katyperry', 'ladygaga', 'TheEllenShow']
#sports = ['Cristiano', 'neymarjr', 'KingJames', 'imVkohli']

celebs_dict = {'elonmusk': 'Elon Musk', 
               'sundarpichai': 'Sundar Pichai',
               'tim_cook': 'Tim Cook',
               'finkd': 'Mark Zuckerberg',
               'JeffBezos': 'Jeff Bezos',
               'satyanadella': 'Satya Nadella',
               'BillGates': 'Bill Gates',
               'ZelenskyyUa': 'Volodymyr Zelenskyy',
               'POTUS': 'Joe Biden',
               'narendramodi': 'Narendra Modi',
               'AOC': 'Alexandria Ocaso-Cortez',
               'GovRonDeSantis': 'Ron DeSantis',
               'BarackObama': 'Barack Obama',
               'KimKardashian': 'Kim Kardashian', 
               'rihanna': 'Rihanna',
               'justinbieber': 'Justin Bieber',
               'katyperry': 'Katy Perry',
               'ladygaga': 'Lady Gaga',
               'TheEllenShow': 'Ellen Degeneres'
              }

# Search definition 
# tweet_count = 1 --> --max-results {tweet_count}
since_date = '2022-01-01'
```

```python
for cat in tqdm([techCEOs, politicians, celebs]):
    if cat == techCEOs:
        category = 'Tech CEO'
    elif cat == politicians:
        category = 'Politician'
    elif cat == celebs:
        category = 'Celebrities'
        
    for user in tqdm(cat):
        #Define a json_file for each user
        json_file = f'{user}_tweets_2022.json'

        # Using OS library to get tweets from selected users and save them as jsons 
        os.system(f'snscrape --jsonl --progress --since {since_date} twitter-search "from:{user}" > ./data/jsons/{json_file}');

        # Reading each json file as a dataframe
        globals()[f'{user}'] = pd.read_json(f'./data/jsons/{json_file}', lines=True)
        globals()[f'{user}']['user_name'] = user
        globals()[f'{user}']['category'] = category   
```

```python
tweets_raw = pd.concat([globals()[f'{user}'] for user in techCEOs + politicians + celebs])
tweets_raw['name'] = tweets_raw['user_name'].map(celebs_dict)
tweets_raw.to_csv('./data/tweets_raw.csv', index=False)
```
