```python
# Import libraries

import pandas as pd
import os
```

```python
# Load raw data from scraping

tweets_raw = pd.read_csv('../data/tweets_raw.csv')
tweets_raw.head()
```

```python
# Data cleansing

# Drop NaN and other problematic data
tweets_dropnas = tweets_raw.dropna(subset=['replyCount']).drop(index=14215).drop(index=14208)

tweets_dropnas.info()
```

```python
tweets_cols1 = tweets_dropnas[['name','user', 'date','category', 'id', 'content','lang', 'sourceLabel',
                           'inReplyToTweetId', 'inReplyToUser', 'mentionedUsers', 'hashtags', 'replyCount']]

tweets_cols_to_num = tweets_dropnas[['retweetCount','likeCount', 'quoteCount']].apply(pd.to_numeric, errors='coerce')

tweets_clean = pd.concat((tweets_cols1, tweets_cols_to_num), axis=1)

tweets_clean.to_csv('../data/tweets_clean.csv', index=False)
```
