```python
import pandas as pd
import datetime as dt
import plotly.express as px
import seaborn as sns

pd.set_option('display.max_columns', None)
```

```python
df_twitter = pd.read_csv('../data/tweets_raw.csv')
df_twitter.head()
```

```python
df_1=df_twitter[['date','replyCount', 'retweetCount', 'likeCount','quoteCount','name']]
df_1['likeCount']=pd.to_numeric(df_1['likeCount'])
df_1['ratio_like_to_reply']=df_1['likeCount']/df_1['replyCount']
df_1['ratio_like_to_retweet']=df_1['likeCount']/df_1['retweetCount']

df_1
```

```python
sns.lineplot(x="date", y="ratio_like_to_reply",
             data=df_1)
```

```python
sns.lineplot(x="date", y="ratio_like_to_retweet",
             data=df_1)
```

```python
sns.lineplot(x="date", y="likeCount",
             data=df_1)
```

```python
sns.lineplot(x="date", y="retweetCount",
             data=df_1)
```

```python
n_days=df_1.groupby(pd.Grouper(key='date',freq='D')).count()
sns.lineplot(x="date", y="replyCount",
             data=n_days)
```

```python
#number of tweets every day/month/week
days=df_1.groupby(pd.Grouper(key='date',freq='D')).sum()
days['ratio_like_to_reply']=days['likeCount']/days['replyCount']
days['ratio_like_to_retweet']=days['likeCount']/days['retweetCount']
days
```

```python
sns.set_theme(style="whitegrid")

sns.color_palette("dark:salmon_r", as_cmap=True)

#cmap = sns.cubehelix_palette(rot=-.2, as_cmap=True)
g = sns.relplot(
    data=days,
    x="date", y="likeCount",
    hue="retweetCount",
    height=8.27, aspect=11.7/8.27
)

g.set(yscale="log")
g.ax.xaxis.grid(True, "minor", linewidth=.25)
g.ax.yaxis.grid(True, "minor", linewidth=.25)
g.despine(left=True, bottom=True)
```

```python
months=df_1.groupby(pd.Grouper(key='date',freq='M')).sum()
months['ratio_like_to_reply']=months['likeCount']/months['replyCount']
months['ratio_like_to_retweet']=months['likeCount']/months['retweetCount']
#sns.set_theme(style="whitegrid")
sns.color_palette("dark:salmon_r", as_cmap=True)

#cmap = sns.cubehelix_palette(rot=-.2, as_cmap=True)
g = sns.relplot(
    data=months,
    x="date", y="likeCount",
    hue="ratio_like_to_reply", size='ratio_like_to_retweet',
    height=8.27, aspect=11.7/8.27
)

g.set(yscale="log")
g.ax.xaxis.grid(True, "minor", linewidth=.25)
g.ax.yaxis.grid(True, "minor", linewidth=.25)
g.despine(left=True, bottom=True)
```
