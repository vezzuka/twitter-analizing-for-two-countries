```python
# Import packages

import os
import pandas as pd
```

```python
# Initialize variables
# Selected users
techCEOs = ['elonmusk', 'sundarpichai', 'tim_cook', 'BillGates', 'finkd', 'JeffBezos', 'satyanadella']
politicians = ['ZelenskyyUa', 'POTUS', 'AOC', 'GovRonDeSantis', 'MedvedevRussiaE', 'MedvedevRussia']
celebs = ['KimKardashian', 'rihanna']

celebs_dict = {'elonmusk': 'Elon Musk', 
               'sundarpichai': 'Sundar Pichai',
               'tim_cook': 'Tim Cook'
              }

# Search definition 
tweet_count = 50
since_date = '2022-01-01'
```

```python
for cat in [techCEOs, politicians, celebs]:
    for user in cat:
        #Define a json_file for each user
        json_file = f'{user}_tweets_2022.json'

        # Using OS library to get tweets from selected users and save them as jsons
        os.system(f'snscrape --jsonl --progress --max-results {tweet_count} --since {since_date} twitter-search "from:{user}" > ./data/jsons/{json_file}');

        # Reading each json file as a dataframe
        globals()[f'{user}'] = pd.read_json(f'./data/jsons/{json_file}', lines=True)
        globals()[f'{user}']['user_name'] = user
        globals()[f'{user}']['category'] = str(cat)   
```

```python
tweets_raw = pd.concat([globals()[f'{user}'] for user in techCEOs + politicians + celebs])
tweets_raw.to_csv('./data/tweets_raw.csv', index=False)
```
